```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from api_keys import g_key
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
cd = pd.read_csv("../output_data/cities.csv")
cd.head()
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
gmaps.configure(api_key=g_key)
```

```python
locations = cd[["Lat", "Lng"]]
humidity = cd["Humidity"]
fig = gmaps.figure()
heat= gmaps.heatmap_layer(locations, weights=humidity, dissipating=False, max_intensity=300, point_radius=5)

fig.add_layer(heat)

fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
fitCity = cd.loc[(cd["Max Temp"] < 72) & (cd["Max Temp"] > 64) & (cd["Wind Speed"] < 10) & (cd["Cloudiness"] == 0)].dropna()
fitCity
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
hdf = fitCity[["City", "Country", "Lat", "Lng"]].copy()
hdf["Hotel"] = ""
hdf
```

```python
hotelParams = {"radius": 5000,"types": "lodging","key": g_key}

for index, row in hdf.iterrows():
    lat = row["Lat"]
    lng = row["Lng"]
    hotelParams["location"] = f"{lat},{lng}"
    
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    name_address = requests.get(base_url, params=hotelParams)
    name_address = name_address.json()
    
    try:
        hdf.loc[index, "Hotel"] = name_address["results"][0]["name"]
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")

hdf
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
info = [info_box_template.format(**row) for index, row in hdf.iterrows()]
locations = hdf[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map
marker_layer = gmaps.marker_layer(locations, info_box_content=info)
fig.add_layer(marker_layer)

# Display figure
fig
```
